```python
from colabdesign import esm_msa
import numpy as np
import jax
import matplotlib.pyplot as plt
```

```python
%shell wget -qnc https://github.com/sokrypton/ColabDesign/blob/v1.0.9/esm_msa/test_msa.npy?raw=true -O test_msa.npy
tokens = np.load('test_msa.npy')[0]
print(tokens.shape)
plt.figure()
plt.imshow(tokens)
plt.show()
```

```python
model, alphabet = esm_msa.pretrained.get_model()
_ = model.init(model.key, tokens)
model.load_params('esm_msa_jax_v1.pkl')
```

```python
result = model(tokens)
print(result.keys())
```

```python
contacts = result['contacts']
plt.figure()
plt.imshow(contacts)
plt.show()
```
